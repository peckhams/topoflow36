# How to Prepare a Country DEM

## Download Source DEM that Contains the Country <a id="data_B"></a>

This notebook has been designed to create a MERIT DEM for any country given its geographic bounding box and the desired spatial resolution (grid cell size).
It assumes that you have the necessary source DEM, in GeoTIFF format, for a larger region that contains the country.  For the DEM, this could be a global DEM, for example.  The source DEM will be clipped to the river basin's geographic bounding box and also resampled to the desired spatial resolution.  The
[<b>MERIT DEM</b>](http://hydro.iis.u-tokyo.ac.jp/~yamadai/MERIT_DEM/)
tiles may be mosaicked and/or clipped to create the source DEM.

## Clip a Source DEM to a Bounding Box and Resample  <a id="clip_dem_B"></a>

Here, we use the TopoFlow <b>regrid</b> utility to <b>clip</b> a DEM for a larger region (which contains the country of interest) to the geographic bounding box for a particular country.  This utility uses the gdal.warp() function in the GDAL Python package.

At the same time, we have the option to <b>resample</b> (via spatial bilinear interpolation) the resulting DEM to a different, coarser spatial resolution.  The source DEM has a grid cell size of 3 arcseconds (roughly 90 meters).  Both the source DEM and new DEM are stored in GeoTIFF format.  Resampling typically causes the bounding box to change slightly.

```python
from topoflow.utils import regrid

### in_nodata = ????
### out_nodata = -9999.0

in_file  = src_dem_dir + src_dem
out_file = out_dir + country + '_rawDEM.tif'

regrid.regrid_geotiff(in_file=in_file, out_file=out_file, 
                   out_bounds=out_bounds,
                   out_xres_sec=out_xres_sec, out_yres_sec=out_yres_sec,
                   ### in_nodata=None, out_nodata=None, 
                   RESAMPLE_ALGO='bilinear', REPORT=True)
```
